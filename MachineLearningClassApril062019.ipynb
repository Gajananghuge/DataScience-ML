Python Class

```python
'''
    Python class Syntax
    class className():
'''
#example class defincation
class Circle(object):
    def __init__(self,radius,color):
        self.radius = radius
        self.color = color
        
    def add_radius(self, r):
        self.radius = self.radius+r
        return(self.radius)
    
    def draw_cricle(self):
        plt.gca().add_patch(plt.Circle((0,0), radius = self.radius, fc = self.color))
        plt.axis('scaled')
        plt.show
```

```python
RedCircle = Circle(10, 'red')
```

```python
dir(RedCircle)
```

```python
RedCircle.color
```

```python
RedCircle.draw_cricle()
```

```python
#change some attributes of objects
RedCircle.radius = 1
RedCircle.radius
```

```python
#inheritanc in Python
class Bird:
    
    def __init__(self):
        print('Bird is ready')
        
    def whoisThis(self):
        print("Bird")
        
    def swim(self):
        print("Swim faster")
        
#child class
class Penguin(Bird):
    def __init__(self):
        super().__init__()
        print("Penguin is ready")
    
    def whoisThis(self):
        print("Penguin")
        
    def run(self):
        print("Run faster")    
```

```python
peggy = Penguin()
peggy.whoisThis()
peggy.swim()
peggy.run()
```
